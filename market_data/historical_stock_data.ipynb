# Historical Stock Market Data
Retrieve daily OHLCV (Open, High, Low, Close, Volume) data using pandas-datareader.

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
ticker = "AAPL"
start = datetime(2021, 1, 1)
end = datetime(2025, 12, 31)
```

```python
df = yf.download(
                tickers=ticker, 
                 interval="1d",
                 start=start, 
                 end=end, 
                 group_by="ticker", 
                 multi_level_index=False)

df.head()
```
